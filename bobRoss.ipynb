```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
filename = '/content/drive/MyDrive/Data Science Class/week 8 9/elements-by-episode.xlsx'
df = pd.read_excel(filename)
```

```python
df.head()
```

```python
percent = (df.groupby('Element')['Included'].sum() / len(df['Episode'].unique())) * 100
percent = percent.sort_values()
percent
```

```python
plt.style.use('seaborn-white')
ax = percent.plot(kind="barh", figsize=(15, 40))
ax.xaxis.set_label_position('top')
for index, value in enumerate(percent):
    plt.text(value, index,
             str(round(value, 2)))
plt.title('Paintings of Bob Ross')
plt.xlabel('Percentage of Paintings Containing Each Element')
plt.ylabel('')
plt.xlim(0, 100)
plt.show();
```
